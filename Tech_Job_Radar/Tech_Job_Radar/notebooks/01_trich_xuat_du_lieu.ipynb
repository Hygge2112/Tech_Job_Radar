```python
import pandas as pd
import os

# --- CẤU HÌNH ĐƯỜNG DẪN ---
file1_path = '../data/ket_qua_review_1900.csv'
file2_path = '../data/reviews_1900_pages_5.csv'
file3_path = '../data/ket_qua_review_1900_full.csv'

# --- 1. ĐỌC VÀ CHUẨN HÓA ---
dfs = [] 

# Xử lý File 1 (Không có địa điểm -> Gán "Toàn quốc")
if os.path.exists(file1_path):
    print(f" Đang đọc {file1_path}...")
    df1 = pd.read_csv(file1_path)
    df1_std = pd.DataFrame()
    df1_std['Ten_Cong_Ty'] = df1['Công ty']
    df1_std['Noi_Dung_Review'] = df1['Ưu điểm'].fillna('') + " " + df1['Nhược điểm'].fillna('')
    df1_std['Khu_Vuc'] = "Toàn quốc" 
    dfs.append(df1_std)

# Xử lý File 2 (CÓ ĐỊA ĐIỂM -> Lấy nguyên gốc)
if os.path.exists(file2_path):
    print(f" Đang đọc {file2_path} (Có địa điểm cụ thể)...")
    df2 = pd.read_csv(file2_path)
    df2_std = pd.DataFrame()
    df2_std['Ten_Cong_Ty'] = df2['Tên công ty']
    df2_std['Noi_Dung_Review'] = (
        df2['Tiêu đề'].fillna('') + " " + 
        df2['Ưu điểm'].fillna('') + " " + 
        df2['Nhược điểm'].fillna('')
    )
    # Lấy chính xác địa điểm (VD: Hà Nội, Hồ Chí Minh...)
    df2_std['Khu_Vuc'] = df2['Địa điểm'].fillna('Toàn quốc')
    dfs.append(df2_std)

# Xử lý File 3 (Không có địa điểm -> Gán "Toàn quốc")
if os.path.exists(file3_path):
    print(f" Đang đọc {file3_path}...")
    df3 = pd.read_csv(file3_path)
    df3_std = pd.DataFrame()
    df3_std['Ten_Cong_Ty'] = df3['Công ty']
    df3_std['Noi_Dung_Review'] = df3['Ưu điểm'].fillna('') + " " + df3['Nhược điểm'].fillna('')
    df3_std['Khu_Vuc'] = "Toàn quốc"
    dfs.append(df3_std)

# --- 2. HỢP NHẤT ---
if dfs:
    df_merged = pd.concat(dfs, ignore_index=True)
    
    output_path = '../data/processed/raw_data.csv'
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    df_merged.to_csv(output_path, index=False)
    
    print("-" * 30)
    print(f"Đã hợp nhất {len(df_merged)} dòng.")
    print(" Dữ liệu từ File 2 sẽ hiển thị địa điểm cụ thể (Hà Nội, HCM...).")
    print(" Dữ liệu từ File 1 & 3 sẽ được gán là 'Toàn quốc'.")
else:
    print(" Không có dữ liệu.")
```
