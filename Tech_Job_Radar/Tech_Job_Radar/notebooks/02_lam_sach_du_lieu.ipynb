```python
import pandas as pd
import re

# ƒê·ªçc file raw_data.csv
# Th·ª≠ nhi·ªÅu b·∫£ng m√£ kh√°c nhau ƒë·ªÉ tr√°nh l·ªói font
try:
    df = pd.read_csv('raw_data.csv', encoding='utf-8')
except UnicodeDecodeError:
    try:
        df = pd.read_csv('raw_data.csv', encoding='utf-16')
    except UnicodeDecodeError:
        df = pd.read_csv('raw_data.csv', encoding='latin-1')

def clean_text_final(text):
    if pd.isna(text):
        return ""
    
    text = str(text)
    
    # 1. X√≥a c√°c k√Ω t·ª± ng√¥i sao (‚òÖ, ‚òÜ,...)
    text = re.sub(r'[‚òÖ‚òÜ‚≠ê‚ú®üåüüå†‚ú°Ô∏è‚ùÇ‚úµ‚ú∂‚ú∑‚ú∏‚úπ‚ú∫‚≠êÔ∏é]', '', text)
    
    # 2. Ch·ªâ gi·ªØ l·∫°i: Ch·ªØ (Ti·∫øng Vi·ªát), S·ªë, Kho·∫£ng tr·∫Øng, D·∫•u ch·∫•m, D·∫•u ph·∫©y
    # Regex n√†y lo·∫°i b·ªè t·∫•t c·∫£ c√°c k√Ω t·ª± ƒë·∫∑c bi·ªát kh√°c (@, #, !, ?,...)
    text = re.sub(r'[^a-zA-Z0-9√Ä√Å√Ç√É√à√â√ä√å√ç√í√ì√î√ï√ô√öƒÇƒêƒ®≈®∆†√†√°√¢√£√®√©√™√¨√≠√≤√≥√¥√µ√π√∫ƒÉƒëƒ©≈©∆°∆ØƒÇ√Ç√ä√î∆†∆Ø∆∞ƒÉ√¢√™√¥∆°∆∞·∫†-·ªπ\s\.,]', '', text)
    
    # 3. X√≥a kho·∫£ng tr·∫Øng th·ª´a (v√≠ d·ª•: "  " th√†nh " ")
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# √Åp d·ª•ng h√†m l√†m s·∫°ch cho c·ªôt T√™n c√¥ng ty v√† N·ªôi dung review
if 'Ten_Cong_Ty' in df.columns:
    df['Ten_Cong_Ty'] = df['Ten_Cong_Ty'].apply(clean_text_final)

if 'Noi_Dung_Review' in df.columns:
    df['Noi_Dung_Review'] = df['Noi_Dung_Review'].apply(clean_text_final)

# L∆∞u k·∫øt qu·∫£ ra file m·ªõi
df.to_csv('clean_data_filtered.csv', index=False, encoding='utf-8-sig')
print("ƒê√£ l·ªçc xong v√† l∆∞u file clean_data_filtered.csv")
```
