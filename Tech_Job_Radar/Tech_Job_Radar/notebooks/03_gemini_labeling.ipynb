```python
import pandas as pd
import os

# --- Cáº¤U HÃŒNH ---
# ÄÆ°á»ng dáº«n Ä‘áº¿n file tá»•ng báº¡n vá»«a gá»™p xong
file_path = r"E:\linhtinh\DangHoc\Khai thÃ¡c dá»¯ liá»‡u\Tech_Job_Radar\Tech_Job_Radar\Tech_Job_Radar\data\processed\MERGED_FINAL_CLEAN.csv"
# (Náº¿u file náº±m á»Ÿ thÆ° má»¥c khÃ¡c, báº¡n hÃ£y Ä‘iá»n Ä‘Æ°á»ng dáº«n Ä‘áº§y Ä‘á»§, vÃ­ dá»¥: r"E:\...\MERGED_FINAL_CLEAN.csv")

# --- Xá»¬ LÃ ---
if os.path.exists(file_path):
    print(f"ğŸ“¥ Äang Ä‘á»c file tá»•ng: {file_path}")
    try:
        df = pd.read_csv(file_path, encoding='utf-8-sig')
    except:
        df = pd.read_csv(file_path, encoding='utf-8')

    print(f"ğŸ“Š Tá»•ng sá»‘ review gá»‘c: {len(df)}")
    print("-" * 60)

    # Danh sÃ¡ch 4 cá»™t phÃ¢n loáº¡i cáº§n tÃ¡ch
    categories = ['Dai_Ngo', 'Van_Hoa', 'Lanh_Dao', 'Phat_Trien']

    for col in categories:
        # TÃªn file Ä‘áº§u ra tÆ°Æ¡ng á»©ng
        output_filename = f"Phan_Loai_{col}.csv"
        
        # --- Bá»˜ Lá»ŒC THÃ”NG MINH ---
        # 1. Chá»‰ láº¥y cÃ¡c cá»™t cáº§n thiáº¿t: TÃªn cÃ´ng ty, Review, vÃ  Cá»™t Ä‘ang xÃ©t
        sub_df = df[['Ten_Cong_Ty', 'Noi_Dung_Review', col]].copy()
        
        # 2. Lá»c bá» cÃ¡c dÃ²ng rÃ¡c (NaN, N/A, Lá»—i, hoáº·c trá»‘ng)
        # Giá»¯ láº¡i nhá»¯ng dÃ²ng cÃ³ Ä‘Ã¡nh giÃ¡ lÃ : Tá»‘t, BÃ¬nh thÆ°á»ng, KÃ©m
        valid_values = ['Tá»‘t', 'BÃ¬nh thÆ°á»ng', 'KÃ©m']
        
        # Chuáº©n hÃ³a vá» chá»¯ thÆ°á»ng Ä‘á»ƒ so sÃ¡nh cho chuáº©n
        mask = sub_df[col].astype(str).str.strip().isin(valid_values)
        
        filtered_df = sub_df[mask]
        
        # LÆ°u ra file riÃªng
        filtered_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
        
        print(f"âœ… ÄÃ£ táº¡o file: {output_filename}")
        print(f"   ğŸ‘‰ Sá»‘ lÆ°á»£ng review há»£p lá»‡: {len(filtered_df)} dÃ²ng (ÄÃ£ lá»c bá» N/A)")
    
    print("-" * 60)
    print("ğŸ‰ HoÃ n táº¥t! Báº¡n Ä‘Ã£ cÃ³ 4 file riÃªng biá»‡t náº±m cÃ¹ng thÆ° má»¥c vá»›i file gá»‘c.")

else:
    print(f"âŒ KhÃ´ng tÃ¬m tháº¥y file táº¡i: {file_path}")
```
